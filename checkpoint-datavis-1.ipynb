```python
import pandas as pd 
data = pd.read_csv('kc_house_data.csv')
data
```

```python
data.isnull().sum()
```

```python
# pas de valeurs nulles 
data.info()
```

```python
#all the data is in numeric format except the date 
days=[]
months=[]
years=[]
for i in data['date']:
    years.append(int(i[0:4]))
    months.append(int(i[4:6]))
    days.append(int(i[6:8]))
    

data['months']=months
data['years']=years
data['days']=days
data.drop(columns=["date"],inplace=True)
data
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
def plot_correlation_map( df ):
    d=df
    d.drop(columns=['id','months','days','zipcode'],inplace=True)
    corr = d.corr()
    

    s , ax = plt.subplots( figsize =( 20 , 20 ) )

    cmap = sb.diverging_palette( 220 , 10 , as_cmap = True )

    s = sb.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
    
plot_correlation_map( data )
```

```python
#the features correlating features > 0.1 
#sqft_living , grade , sqft_above , sqft_living15 , bathrooms , view ,sqft_basement ,  
#bedrooms , lat , waterfront , floors , yr_renovated
```

```python
x=data[['price']]
y=data[['sqft_living' , 'grade' , 'sqft_above' , 'sqft_living15' , 'bathrooms' , 'view' ,'sqft_basement' ,  
'bedrooms' , 'lat' , 'waterfront' , 'floors' , 'yr_renovated']]
x
```

```python
 from sklearn.model_selection import train_test_split
x_train,x_test ,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=30)
```

```python
y_=y_train['sqft_living']
g=plt.scatter(x_train,y_)
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics

model=LinearRegression()   #build linear regression model
model.fit(x_train,y_)  #fitting the training data
predicted=model.predict(x_test) #testing our modelâ€™s performance

print ("a= ",model.coef_, "  b= ",model.intercept_)

plt.scatter(x_train,y_,color="r")
plt.plot(x_test,predicted,color="k")
plt.show()


```

```python
# R squared is approximatively equal to 0.5 so only half of the output can be explained by the input model
print("MSE", mean_squared_error(y_test[['sqft_living']],predicted))
print("R squared", metrics.r2_score(y_test[['sqft_living']],predicted))

```
